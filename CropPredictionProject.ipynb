```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import pickle
```

```python
dataset = pd.read_csv("/content/drive/MyDrive/crop.csv")
```

```python
dataset
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset['season'].unique()
```

```python
encoder = OrdinalEncoder()
```

```python
dataset['season'] = encoder.fit_transform(dataset[['season']])
```

```python
dataset['season'].unique()
```

```python
Y = dataset['label']
```

```python
X = dataset[['temperature','humidity','ph','water availability','season']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
x_test
```

Data Modelling

Decision Tree Classifier (DTC)

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
with open("dtc_model.pkl", "wb") as file:
    pickle.dump(model, file)
```

```python
model_pred = model.predict(x_test)
```

```python
model_pred
```

```python
print(classification_report(y_test, model_pred))
```
